```python
import numpy as np
import matplotlib.pyplot as plt
```

SISTEMA 1: SOLUÇÃO ÚNICA

```python
# Sistema 1: Solução única (Plotagem)

x = np.linspace(-10,10)

plt.plot(x, 3 - 2*x, color = 'green')
plt.plot(x, (2 + x)/3, color = 'blue')
plt.title("Sistema 1: Solução única")
plt.plot()
```

```python
# Resolução do sistema 1, de solução única

a = np.array([[2,1],[1,-3]])
b = np.array([3,-2])
c = np.linalg.solve(a,b)

print(c)
```

SISTEMA 2: INFINITAS SOLUÇÕES

```python
# Sistema 2: Infinitas soluções (Plotagem)

x = np.linspace(-10,10)

plt.plot(x, 3 - 2*x, color = 'green')
plt.plot(x, (6 + -4*x)/2, color = 'blue')
plt.title("Sistema 3: Infinitas soluções")
plt.plot()
```

```python
# Resolução do sistema 2, de infinitas soluções

a = np.array([[2,1],[4,2]])
b = np.array([3,6])
c = np.linalg.solve(a,b)

print(c)
```

SISTEMA 3: NENHUMA SOLUÇÃO

```python
# Sistema 3: Nenhuma solução (Plotagem)

x = np.linspace(-10,10)

plt.plot(x, 3 - 2*x, color = 'green')
plt.plot(x, (2 + -4*x)/2, color = 'blue')
plt.title("Sistema 3: Nenhuma solução")
plt.show()
```

```python
# Resolução do sistema 3, de nenhuma solução

a = np.array([[2,1],[4,2]])
b = np.array([3,2])
c = np.linalg.solve(a,b)

print(c)
```

PRODUTO DE UM VETOR POR UMA MATRIZ

```python
# Define a função plotVet, que exibe graficamente até dois vetores

def plotVet(A,b,c):
    
    # Plota o eixo cartesiano
    x = np.linspace(-2,11)
    y = np.linspace(-5,5)
    plt.plot(x,x-x,color = 'black')
    plt.plot(y-y,y,color = 'black')
    
    # Plota o vetor b
    if b != 'NULL':
        bx = np.linspace(0,b[0])
        by = np.linspace(0,b[1])
        plt.plot(bx,by,color='green')
    
    # Plota o vertor c
    if c != 'NULL':
        cx = np.linspace(0,c[0])
        cy = np.linspace(0,c[1])
        plt.plot(cx,cy,color='blue')
                
    # Desenha o gráfico dos vetores originais
    plt.title('GRÁFICO DOS VETORES ORIGINAIS') 
    plt.show()
    
    # Efetua o produto dos vetores pela matriz A e plota os gráficos resultantes
    if b != 'NULL' or c != 'NULL':
        
        # Plota o eixo cartesiano
        x = np.linspace(-2,11)
        y = np.linspace(-5,5)
        plt.plot(x,x-x,color = 'black')
        plt.plot(y-y,y,color = 'black')
                
        # Calcula o vetor resultante do produto de A por b e plota o gráfico     
        if b != 'NULL':
            u = np.matmul(A.data,b)
            ux = np.linspace(0,u[0])
            uy = np.linspace(0,u[1])
            plt.plot(ux,uy,color = 'green')
        
        # Calcula o vetor resultante do produto de A por c
        if c != 'NULL':
            v = np.matmul(A.data,c)
            vx = np.linspace(0,v[0])
            vy = np.linspace(0,v[1])
            plt.plot(vx,vy, color = 'blue')
            
        # Desenha os gráficos resultantes
        plt.title('GRÁFICOS RESULTANTES APÓS O PRODUTO')
        plt.show()
        
    # Printa os vetores resultantes
    if b != 'NULL': print('b =', u)
    if c != 'NULL': print('c =', v)
```

```python
# Define os parâmetros passados para a função        
A  = np.matrix([[2, 1],[1, -3]]) 
b = [1,1]
c = [2,-1]

# Plota os gráficos resultantes
plotVet(A,b,c)
```

VETOR RESULTANTE DA SOMA DE DOIS VETORES MULTIPLICADOS POR DOIS ESCALARES

```python
# Define a função plotSom, que exibe graficamente a soma de dois vetores

def plotSom(m,n,b,c):
    
    # Plota o eixo cartesiano
    x = np.linspace(-2,11)
    y = np.linspace(-5,5)
    plt.plot(x,x-x,color = 'black')
    plt.plot(y-y,y,color = 'black')
    
    # Plota o vetor b
    bx = np.linspace(0,b[0])
    by = np.linspace(0,b[1])
    plt.plot(bx,by,color='green')
    
    # Plota o vertor c
    cx = np.linspace(0,c[0])
    cy = np.linspace(0,c[1])
    plt.plot(cx,cy,color='blue')
                
    # Desenha o gráfico dos vetores originais
    plt.title('GRÁFICO DOS VETORES ORIGINAIS') 
    plt.show()
    
    # Efetua o produto dos vetores pelos escalares m e n e calcula a soma dos vetores resultantes

    # Plota o eixo cartesiano
    x = np.linspace(-2,11)
    y = np.linspace(-5,5)
    plt.plot(x,x-x,color = 'black')
    plt.plot(y-y,y,color = 'black')
                
    # Calcula o vetor resultante do produto do escalar m por b e plota o gráfico     
    u = [m*b[0],m*b[1]]
    ux = np.linspace(0,u[0])
    uy = np.linspace(0,u[1])
    plt.plot(ux,uy,color = 'green')
        
    # Calcula o vetor resultante do produto do escalar n por c e plota o gráfico
    v = [n*c[0],n*c[1]]
    vx = np.linspace(0,v[0])
    vy = np.linspace(0,v[1])
    plt.plot(vx,vy, color = 'blue')
    
    # Calcula o vetor soma s = b + c
    s = [u[0]+v[0],u[1]+v[1]]
    sx = np.linspace(0,s[0])
    sy = np.linspace(0,s[1])
    plt.plot(sx,sy, color = 'red')
            
    # Desenha os gráficos resultantes
    plt.title('GRÁFICO DO VETOR SOMA')
    plt.show()
    
    # Printa os vetores resultantes
    print('b =', u)
    print('c =', v)
    print('s =', s)
```

```python
# Define os parâmetros passados para a função        
m = 2
n = 1
b = [2,1]
c = [1,-3]

# Plota os gráficos resultantes
plotSom(m,n,b,c)
```
